```python
import json

dataset_type = 'shp'

file1 = f"gen_responses_intervented_vicuna_7B_{dataset_type}.json"
file2 = f"gen_responses_vicuna_7B_{dataset_type}.json"

with open(file1, "r", encoding="utf-8") as f1, open(file2, "r", encoding="utf-8") as f2:
    data1 = json.load(f1)
    data2 = json.load(f2)

map1 = {item["prompt"]: item["result"] for item in data1}
map2 = {item["prompt"]: item["result"] for item in data2}

diff_items = []
for prompt in map1:
    if prompt in map2 and map1[prompt] != map2[prompt]:
        diff_items.append({
            "prompt": prompt,
            "result_file1": map1[prompt],
            "result_file2": map2[prompt]
        })
```
